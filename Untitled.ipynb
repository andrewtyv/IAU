```python
def null_values(df):

    null_counts = df.isnull().sum()  # Count null values in each column

    print("null values in each column")
# Create a DataFrame to present the results
    null_counts_df = pd.DataFrame(null_counts, columns=['Null Count'])  # Create a DataFrame for better visualization
    null_counts_df['Non-null Count'] = df.shape[0] - null_counts_df['Null Count']  # Calculate non-null count
    null_counts_df['Percentage of Nulls'] = (null_counts_df['Null Count'] / df.shape[0]) * 100  # Calculate the percentage of null values
    print(null_counts_df)
```

```python
def generate_outliers():
    outliers = pd.DataFrame()
    for column in df_merged.select_dtypes(include=['float64', 'int64']).columns:
        if (column == 'imei' or column =='mwra'):
            continue
        Q1 = df_merged[column].quantile(0.25)
        Q3 = df_merged[column].quantile(0.75)
        IQR = Q3 - Q1  
    
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
    
        outliers[column] = df_merged[column][(df_merged[column] < lower_bound) | (df_merged[column] > upper_bound)]
    return outliers
```

```python

def replace_outliers_with_boundaries(df, column):
    lower_bound = df[column].quantile(0.25)  
    upper_bound = df[column].quantile(0.75) 
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
    
```

```python
columns_to_replace = ['Adj Close','Close','High','Low','Open','Volume','Currentprice','Marketcap','Ebitda','Revenuegrowth','Fulltimeemployees']

for column in columns_to_replace:
    df_merged = replace_outliers_with_boundaries(df_merged, column)
```

```python

def make_stationary(series, significance_level=0.05):
    
    differencing_count = 0
    adf_test_result = adfuller(series.dropna())
    p_value = adf_test_result[1]
    
    print(f"Initial P-value: {p_value}")
    
    while p_value > significance_level:
        differencing_count += 1
        series = series.diff()
        adf_test_result = adfuller(series.dropna())
        p_value = adf_test_result[1]
        print(f"After {differencing_count} differencing(s), P-value: {p_value}")
        
        if differencing_count > 10:  # Safety check to prevent infinite loops
            print("Stopping after 10 differences. Series might not be stationary.")
            break
    
    if p_value <= significance_level:
        print(f"The series became stationary after {differencing_count} differencing(s).")
    else:
        print("The series did not become stationary within the allowed iterations.")
    
    return series.dropna(), differencing_count

```

```python
df_merged['Date'] = pd.to_datetime(df_merged['Date'], errors='coerce')


df_merged.set_index('Date', inplace=True)

series = df_merged['S&P500']

stationary_series, differencing_count = make_stationary(series)


p, d, q = 1, differencing_count, 1

model = ARIMA(stationary_series, order=(p, d, q))
model_fit = model.fit()

print(model_fit.summary())

forecast = model_fit.forecast(steps=10)  
print("Forecasted values:", forecast)

plt.figure(figsize=(12, 6))
plt.plot(stationary_series, label='Actual Data')
plt.plot(pd.date_range(stationary_series.index[-1], periods=11, freq='D')[1:], forecast, label='Forecast', color='red')
plt.title("ARIMA Forecast for S&P500")
plt.legend()
plt.show()
```
